# Project Title

E-Commerce Transaction Analytics: Revenue, Customer Trends, and Operational Insights (2022)

# Domain

This project falls under the Retail and E-Commerce domain, focusing on analyzing online transaction data to understand customer purchasing behavior, sales performance, delivery efficiency, and return patterns.

# Objective

1. To analyze e-commerce transaction data from 2022 to identify patterns and trends in customer purchasing behavior.

2. To clean, preprocess, and transform raw transaction data for accurate exploratory data analysis (EDA).

3. To evaluate product category performance, payment method preferences, and return behavior.

4. To examine the relationship between delivery time, customer ratings, and return rates.

5. To generate actionable business insights that can support data-driven decision-making.

# Outcome

This project provides a comprehensive understanding of sales performance and customer behavior in an e-commerce environment. The analysis uncovers trends related to revenue generation, product popularity, customer satisfaction, and operational efficiency. The findings can assist businesses in optimizing pricing strategies, improving delivery performance, and reducing product return rates to enhance overall profitability.

#Dataset Information

Source: https://www.kaggle.com/datasets/dastgeerjutt/e-commerce-customer-behavior

Year / Timeline: The dataset contains e-commerce transaction records from 2022, representing customer purchase behavior over the course of the year.

Dataset Description:

The dataset consists of 5000 customer transaction records from an e-commerce platform. Each record represents an individual order and contains information related to customer demographics, product category, payment method, order value, delivery time, customer ratings, and return status.

The dataset is useful for analyzing customer purchasing patterns, evaluating product performance, understanding payment preferences, and identifying factors that influence customer satisfaction and product returns.

# Type of Analysis

Descriptive Analysis :

The project primarily falls under Descriptive Analysis because it summarizes and interprets the e-commerce transaction dataset using statistical measures and visualizations. The analysis focuses on understanding customer demographics, product category performance, payment preferences, order values, and delivery times. Charts and summary statistics are used to describe patterns and trends in the data.

Diagnostic Analysis :

 The project also includes Diagnostic Analysis as it examines relationships between different variables to identify reasons behind observed patterns. For example, the analysis investigates how delivery time affects customer ratings, how product categories influence return rates, and how customer demographics impact purchasing behavior. This helps explain why certain trends occur in the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
#Product Category Distribution
plt.figure(figsize=(8,5))
sns.countplot(x='product_category', data=df)
plt.xticks(rotation=45)
plt.title("Product Category Distribution")
plt.show()
```

Interpretation of the Above Chart

Explanation

This chart represents the distribution of different product categories in the dataset. It shows how many orders belong to each category such as Home & Living, Fashion, Sports, Electronics, Groceries, and Beauty.

What is it saying?

The chart indicates that all product categories have a relatively similar number of orders, with categories like Fashion, Sports, and Beauty showing slightly higher counts compared to others.

What features you used?

The visualization was created using Seaborn's countplot() function to display the frequency of each product category.

From this what are we showing?

This analysis helps understand the distribution of customer purchases across different product categories. which can help businesses identify popular categories and plan inventory or marketing strategies accordingly.

```python
#Revenue by Product Category
plt.figure(figsize=(8,5))
sns.barplot(x='product_category', y='order_value_usd', data=df)
plt.xticks(rotation=45)
plt.title("Average Order Value by Product Category")
plt.show()
```

Interpretation of the Above Chart

Explanation:

This bar chart illustrates the average order value for each product category in the dataset. It compares how much customers spend on average across categories such as Home & Living, Fashion, Sports, Electronics, Groceries, and Beauty.

What is it saying?

The chart shows that categories like Fashion and Groceries have slightly higher average order values, while Electronics and Beauty have slightly lower average order values. However, the overall spending across categories appears fairly similar.

What features you used?

The visualization was created using Seaborn's barplot() function, which calculates and displays the average order value for each category.


From this what are we showing?

This analysis highlights how customer spending varies across different product categories, helping businesses identify categories that generate higher revenue per order and focus marketing or promotional efforts accordingly.

```python
#Monthly Revenue Trend
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

df['month'] = df['order_date'].dt.month

monthly_sales = df.groupby('month')['order_value_usd'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

What is this chart?

This is a line chart showing how revenue changes over 12 months (Month 1 to Month 12).

X-axis → Months (1–12)


Y-axis → Revenue (amount earned)

**What is it saying?**
(Key Insights)

1. Highest Revenue

  Around Month 5 (333,000) → This is the peak performance.

2. Lowest Revenue
Around Month 11 (287,000) → This is the lowest point.

3. Fluctuating Trend

 . Revenue is not stable.

 . It goes up and down throughout the year.

4. Strong Start

Month 1 starts high (330,000)

Big drop in Month 2
